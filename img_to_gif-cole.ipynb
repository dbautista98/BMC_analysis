```python
from PIL import Image
import glob
```

```python
def onion_path(onion, flow, num):
    path = "../data/onion_%s/Onion_%s_20x-%s*"%(onion, flow, num)
    return glob.glob(path)[0]
onion_path("4", "bulk", "18")
```

A quick program to convert image data to gif format following example from https://pythonprogramming.altervista.org/png-to-gif/

```python
inpath = onion_path("4", "bulk", "18")
outpath = "/mnt/c/Users/coler/OneDrive/Documents/BMC/data/movies/onionflow.gif"
```

```python
img = glob.glob(inpath+"/*.bmp")

frames = []
for i in img:
    new_frame = Image.open(i)
    frames.append(new_frame)

frames[0].save(outpath, format="GIF",
               append_images=frames[1:],
               save_all=True,
               duration=300, loop=0)
```
